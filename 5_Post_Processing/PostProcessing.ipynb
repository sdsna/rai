```python
import geopandas as gpd
import pandas as pd
import os
import numpy as np
```

```python
df = pd.read_excel('/Users/guilhermeiablonovski/Dropbox (SDSN)/SDG Geospatial Indicators Project/sdg-roadinfrastructure/_processed_data/2023_RAI.xlsx')
```

```python
df
```

```python
rai_region = df.groupby('WB region').apply(lambda x: np.average(x['RAI (GDP corrected)'], weights=x['Total Rural Population (SMOD/WorldPop)'])).rename("RAI_region")
#sedac_region = df.groupby('WB region').apply(lambda x: np.average(x['RAI (SEDAC by SDSN)'], weights=x['Total Rural Population (SMOD/WorldPop)'])).rename("SEDAC_region")
pop2015 = df.groupby('WB region').apply(lambda x: np.sum(x['Total Rural Population (SMOD/WorldPop)'])).rename("Rural Pop")
weighted_region_wb = pd.concat([rai_region,pop2015], axis=1)
```

```python
weighted_region_wb
```

```python
rai_region = df.groupby('region').apply(lambda x: np.average(x['RAI (GDP corrected)'], weights=x['Total Rural Population (SMOD/WorldPop)'])).rename("RAI_region")
#sedac_region = df.groupby('region').apply(lambda x: np.average(x['RAI (SEDAC by SDSN)'], weights=x['Total Rural Population (SMOD/WorldPop)'])).rename("SEDAC_region")
pop2015 = df.groupby('region').apply(lambda x: np.sum(x['Total Rural Population (SMOD/WorldPop)'])).rename("Rural Pop")
weighted_region = pd.concat([rai_region,pop2015], axis=1)
```

```python
weighted_region
```

```python
rai_region = df.groupby('WB income group').apply(lambda x: np.average(x['RAI (GDP corrected)'], weights=x['Total Rural Population (SMOD/WorldPop)'])).rename("RAI_region")
sedac_region = df.groupby('WB income group').apply(lambda x: np.average(x['RAI (SEDAC by SDSN)'], weights=x['Total Rural Population (SMOD/WorldPop)'])).rename("SEDAC_region")
pop2015 = df.groupby('WB income group').apply(lambda x: np.sum(x['Total Rural Population (SMOD/WorldPop)'])).rename("Rural Pop")
weighted_income = pd.concat([rai_region,sedac_region,pop2015], axis=1)
```
